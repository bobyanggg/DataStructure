```python
class Vertex:
    def __init__(self,n):
        self.name =n
        self.neighbors = list()
        
        self.discovery = 0
        self.finish = 0
        self.color = 'black'
    
    def add_neighbor(self,v):
        nset = set(self.neighbors)
        if v not in nset:
            self.neighbors.append(v)
            self.neighbors.sort()

class Graph:
    vertices = {}
    time = 0
    
    def add_vertex(self,vertex):
        if isinstance(vertex,Vertex) and vertex.name not in self.vertices:
            self.vertices[vertex.name] = vertex
            return True
        
        else:
            return False
    
    def add_edge (self, u, v):
        if u in self.vertices and v in self.vertices:
            for key, value in self.vertices.items():
                
                if key == u:
                    value.add_neighbor(v)
                
                if key == v:
                    value.add_neighbor(u)
            return True
        
        else:
            return False
    
    def print_graph(self):
        for key in sorted(list(self.vertices)):
            print (key + str(self.vertices[key].neighbors)+str(self.vertices[key].discovery)+"/"+str(self.vertices[key].finish))
            
    def _dfs(self, vertex):
        global time
        vertex.color = 'red'
        vertex.discovery = time
        time += 1
        for v in vertex.neighbors:
            if self.vertices[v].color == 'black':
                self._dfs(self.vertices[v])
        vertex.color = 'blue'
        vertex.finish = time
        time += 1
        
    def dfs (self, vertex):
        global time
        time = 1
        self._dfs(vertex)

g = Graph()

a = Vertex('A')
g.add_vertex(a)
g.add_vertex(Vertex('B'))

for i in range (ord('A'), ord('K')):
    g.add_vertex(Vertex(chr(i)))
    
edges = ['AB', 'AE', 'BF', 'CG', 'DE', 'DH', 'EH', 'FG', 'FJ', 'GJ', 'HI']

for edge in edges:
    g.add_edge(edge[:1], edge[1:])
    
g.dfs(a)
g.print_graph()
    
    
    
    
    
```

```python
#BFS
class Vertex:
    def __init__(self,n):
        self.name =n
        self.neighbors = list()
        
        self.distance =0
        
        self.discovery = 0
        self.finish = 0
        self.color = 'black'
    
    def add_neighbor(self,v):
        nset = set(self.neighbors)
        if v not in nset:
            self.neighbors.append(v)
            self.neighbors.sort()

class Graph:
    vertices = {}
    time = 0
    
    def add_vertex(self,vertex):
        if isinstance(vertex,Vertex) and vertex.name not in self.vertices:
            self.vertices[vertex.name] = vertex
            return True
        
        else:
            return False
    
    def add_edge (self, u, v):
        if u in self.vertices and v in self.vertices:
            for key, value in self.vertices.items():
                
                if key == u:
                    value.add_neighbor(v)
                
                if key == v:
                    value.add_neighbor(u)
            return True
        
        else:
            return False
    
    def print_graph(self):
        for key in sorted(list(self.vertices)):
            print (key + str(self.vertices[key].neighbors)+" "+str(self.vertices[key].distance))
            
    def bfs(self, vert):
        q = list()
        vert.distance = 0
        vert.color = 'red'
        #1
        for v in vert.neighbors:
            self.vertices[v].distance = vert.distance + 1
            q.append(v)
        #234    
        while len(q) > 0:
            u = q.pop(0)
            
            
            node_u = self.vertices[u]
            node_u.color = 'red'
            
            for v in node_u.neighbors:
                node_v = self.vertices[v]
                if node_v.color == 'black':
                    q.append(v)
                    node_v.color = 'red'
                    node_v.distance = node_u.distance + 1
            
        
g = Graph()

a = Vertex('A')
g.add_vertex(a)
g.add_vertex(Vertex('B'))

for i in range (ord('A'), ord('K')):
    g.add_vertex(Vertex(chr(i)))
    
edges = ['AB', 'AE', 'BF', 'CG', 'DE', 'DH', 'EH', 'FG', 'FJ', 'GJ', 'HI']

for edge in edges:
    g.add_edge(edge[:1], edge[1:])
    
g.bfs(a)
g.print_graph()
    
    
    
    
    
```

```python
#Bellman-Ford
class Vertex:
    def __init__(self,n):
        self.name =n
        self.neighbors = list()
        self.distance = 9999999
        
        self.discovery = 0
        self.finish = 0
        self.predecessor = None
        self.color = 'black'
    
    def add_neighbor(self,v):
        nset = set(self.neighbors)
        if v not in nset:
            self.neighbors.append(v)
            self.neighbors.sort()
            
class Edge:
    def __init__(self, name, weight):
        self.name = name
        self.weight = weight
        self.startVertex = self.name[:1]
        self.targetVertex = self.name[1:]
    

        
class Graph:
    vertices = {}
    edges = {}
    time = 0
    
    def add_vertex(self,vertex):
        if isinstance(vertex,Vertex) and vertex.name not in self.vertices:
            self.vertices[vertex.name] = vertex
            return True
        
        else:
            return False
    
    def add_edge (self, edge):
        if isinstance(edge,Edge):
            self.edges[edge.name] = edge
            return True
        
        else:
            return False
    
    def print_graph(self):
        for key in sorted(list(self.vertices)):
            print (key +'->'+ str(self.vertices[key].distance)+" "+str(self.vertices[key].predecessor))
            
    def _dfs(self, vertex):
        global time
        vertex.color = 'red'
        vertex.discovery = time
        time += 1
        for v in vertex.neighbors:
            if self.vertices[v].color == 'black':
                self._dfs(self.vertices[v])
        vertex.color = 'blue'
        vertex.finish = time
        time += 1
        
    def dfs (self, vertex):
        global time
        time = 1
        self._dfs(vertex)

    
    def cal_shortest_path(self, firstVertex):
        
        firstVertex.distance = 0
        
        for i in range(0,len(self.vertices)):
         
            for key in self.edges:
                
                e = self.edges[key]
                
                u = self.vertices[e.startVertex]
                v = self.vertices[e.targetVertex]
                
                newDistance = u.distance + e.weight
                new2Distance = v.distance + e.weight
            
                if newDistance < v.distance:
                    v.distance = newDistance
                    v.predecessor = e.startVertex
                    self.vertices[e.targetVertex]=v
                   
            
                    
       
        for edge in self.edges:
            if v.distance + e.weight < u.distance:
                print ('Negative Cycle')
            
            

g = Graph()

a = Vertex('A')
g.add_vertex(a)
g.add_vertex(Vertex('B'))
for i in range (ord('A'), ord('G')):
    g.add_vertex(Vertex(chr(i)))
    

g.add_edge(Edge('AB',7))
g.add_edge(Edge('AE',14))
g.add_edge(Edge('AF',9))
g.add_edge(Edge('EF',2))
g.add_edge(Edge('ED',9))
g.add_edge(Edge('FB',10))
g.add_edge(Edge('FC',11))
g.add_edge(Edge('DC',6))
g.add_edge(Edge('BC',15))
g.add_edge(Edge('BA',7))
g.add_edge(Edge('EA',14))
g.add_edge(Edge('FA',9))
g.add_edge(Edge('FE',2))
g.add_edge(Edge('ED',9))
g.add_edge(Edge('BF',10))
g.add_edge(Edge('CF',11))
g.add_edge(Edge('CD',6))
g.add_edge(Edge('CB',15))


g.cal_shortest_path(a)






    
#g.dfs(a)
g.print_graph()

    

    
    
    
    
```

```python
#Dijkstra
class Vertex:
    def __init__(self,n):
        self.name =n
        self.neighbors = {}
        self.distance = 9999999
        
        self.discovery = 0
        self.finish = 0
        self.predecessor = None
        self.color = 'black'
    
    def add_neighbor(self,v,weight):
        nset = set(self.neighbors)
        if v not in nset:
            self.neighbors[v]=weight
            

            
class Edge:
    def __init__(self, name, weight):
        self.name = name
        self.weight = weight
        self.startVertex = self.name[:1]
        self.targetVertex = self.name[1:]   

        
class Graph:
    vertices = {}
    edges = {}
    time = 0           
    path = []
   
    
    def add_vertex(self,vertex):
        if isinstance(vertex,Vertex) and vertex.name not in self.vertices:
            self.vertices[vertex.name] = vertex
            return True
        
        else:
            return False
    
    def add_edge (self, u, v, weight):
        if u in self.vertices and v in self.vertices:
            for key, value in self.vertices.items():
                
                if key == u:
                    value.add_neighbor(v,weight)
                    
                if key == v:
                    value.add_neighbor(u,weight)
                
                
            return True
        
        else:
            return False
        
    
    
    def print_graph(self):
        for key in sorted(list(self.vertices)):
            print (key +'->'+ str(self.vertices[key].distance)+" "+str(self.vertices[key].predecessor))
            
    def _dfs(self, vertex):
        global time
        vertex.color = 'red'
        vertex.discovery = time
        time += 1
        for v in vertex.neighbors:
            if self.vertices[v].color == 'black':
                self._dfs(self.vertices[v])
        vertex.color = 'blue'
        vertex.finish = time
        time += 1
        
    def dfs (self, vertex):
        global time
        time = 1
        self._dfs(vertex)

    
    def cal_shortest_path(self, firstVertex):
        
        firstVertex.distance = 0
        
        for i in range(0,len(self.vertices)):
         
            for key in list(self.edges):
                
                e = self.edges[key]
                
                u = self.vertices[e.startVertex]
                v = self.vertices[e.targetVertex]
                
                newDistance = u.distance + e.weight
                new2Distance = v.distance + e.weight
            
                if newDistance < v.distance:
                    v.distance = newDistance
                    v.predecessor = e.startVertex
                    self.vertices[e.targetVertex]=v
            
                    
       
        for edge in self.edges:
            if v.distance + e.weight < u.distance:
                print ('Negative Cycle')
              
    def Dijkstra (self,start,goal):
        
        unseenNodes = {}
        path = []

        self.vertices[start].distance = 0
        
        for key in self.vertices:
            unseenNodes[key] = self.vertices[key]
            
        
        while unseenNodes:
            
            minNode = None
            #find minNode
            for key in unseenNodes:
                
                
                if minNode is None:
                    minNode = unseenNodes[key]
                
                elif unseenNodes[key].distance < minNode.distance:
                    minNode = unseenNodes[key]
            
            #find shortest path
            for child in minNode.neighbors:
                
                weight = minNode.neighbors[child]

                if weight + minNode.distance < self.vertices[child].distance:
                    self.vertices[child].distance = weight + minNode.distance
                    #string for predecessor
                    self.vertices[child].predecessor = minNode.name
            unseenNodes.pop(minNode.name)
            
        currNode = self.vertices[goal]
        while currNode.name != start:
            try:
                path.insert(0,currNode.name)
                currNode = self.vertices[currNode.predecessor]
            
            except KeyError:
                print('Not reachable')
                break
        path.insert(0,start)
        print ('Path =', str(path))

 
g = Graph()

a = Vertex('A')
g.add_vertex(a)
g.add_vertex(Vertex('B'))
for i in range (ord('A'), ord('G')):
    g.add_vertex(Vertex(chr(i)))
    
g.add_edge('A','B',7)
g.add_edge('A','F',9)
g.add_edge('A','E',14)
g.add_edge('E','F',2)
g.add_edge('E','D',9)
g.add_edge('F','B',10)
g.add_edge('F','C',11)
g.add_edge('D','C',6)
g.add_edge('B','C',15)

g.Dijkstra('A','D')
            
g.print_graph()
        
            
            





    
    
```
